# 问题分解 
### 问题
● 如何将我的数据和程序构造成并行工作？
### 目标
● 理解并行算法和数据的主要策略。

● 了解不同的分解技术。

● 看看如何在一个真正的问题上使用分解

设计并行程序的第一个步骤是将问题分解为可分配到多个任务的离散“块”，以便能够同时处理问题。这就是所谓的分解或分区。分解算法有两种主要方法：域分解和功能分解。

# 域分解
在这种类型的分区中，与问题相关联的数据被分解。然后，每个并行任务都在数据的一部分上工作。

域分解的关键特征是：

● 数据被分割成相同大小的片段并映射到不同的处理器中

● 处理器只处理分配给它的数据

● 在必要时与其他处理器进行通信

有许多不同的方法来划分数据。

# 功能分解
对于函数分解，重点是要执行的计算，而不是由计算所操作的数据。这个问题是根据必须完成的工作来分解的。然后，每个任务执行一部分总体工作。

功能分解的关键特征是：

● 当数据片段需要不同的处理时间时使用

● 性能受最慢的进程限制

● 程序分解成许多小任务

● 当它们可用时分配给处理器的任务

● 在主/从范例中实现

功能分解的例子

● 从有限元网格中进行表面重建

● 搜索图像或数据库
